## imputation missing technique 

# mean value imputationn

```python
sns.displot(df['age'])
```

```python
sns.histplot(df['age'],kde = True)
```

```python
df['age mean'] = df['age'].fillna(df['age'].mean())
```

```python
df[['age mean', 'age']] # mean imputation works well when we have normally distributed data
```

```python
# median value imputatio - if we have outliers in the dataset
df['age median'] = df['age'].fillna(df['age'].median())
```

```python
df[['age median', 'age']]
```
